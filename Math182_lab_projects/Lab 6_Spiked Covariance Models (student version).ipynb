# Lab 6: Spiked Covariance Models#


We now want to understand more general “Wishart” matrices corresponding to spiked covariance models,
of the form $W=\frac{1}{N}X^\top CX$,where $C=I+D_r$. 
Let's assume for now $r=1$, $m=1000$ and $N=500$.
In the following we simulate $W$, with $[D_r]_{11}=\gamma=1$. 

```python
m=1000
N=500
gamma=1
X=np.random.normal(0,1,(m,N))
d=np.zeros(m) #setting up the diagonal of D
d[0]=gamma 
C=np.eye(m)+np.diag(d)
W=1/N*np.dot(np.dot(np.transpose(X),C),X)
```

Compute its eigenvalues and plot their histogram. Superimpose the approriate Marcenko–Pastur density.

```python
#student ###1###
eigvals, eigvecs = np.linalg.eig(W)
plt.hist(eigvals,density=True,bins=25)
x = np.arange(0,4,0.001)  #some random value for x axis
y = 1/2*(np.sqrt((x-2*(1-np.sqrt(1/2))**2)*(2*(1+np.sqrt(1/2))**2-x)))/(2*np.pi*x/2) #y=f(x)
h = plt.plot(x, y, lw=2) 
#plt.axis([0,4,0,2.5]) #zoom in to see better
plt.show()
```
